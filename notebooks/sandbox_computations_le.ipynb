## Computations from W.Thiery(2024) presentation "Will you live an unprecedented life ?"



```python
E = 0.4633 # [GtCO2_eq]
TCRE = 0.5*10**(-3) # [C/GtCO2_eq]
exposure_slope = 10.4 # [heatwaves/C]
cohort_size = 131702400 # [children]

GMT_due_to_emissions = E*TCRE
heatwaves = E*TCRE*exposure_slope
heatwaves_children = E*TCRE*exposure_slope*cohort_size

print("Additional GMT due to emissions = {} C".format(GMT_due_to_emissions))
print("Number of additionnal heatwaves across lifetime = ",heatwaves)
print("Number of heatwaves and children = ",heatwaves_children)
```
